```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data  = [[9,0],[8,0],[7,0],[6,0],[5,1],[4,1],[2,0],[1,1]]
# [3,1],[2,1],
```

```python
df = pd.DataFrame(data, columns = ['Hours_of_study' , 'pass/fail'])
```

```python
df.to_csv('student.csv')
```

```python
plt.scatter(df['Hours_of_study'],df['pass/fail'])
```

```python
from sklearn.linear_model import LinearRegression
```

```python
linear = LinearRegression()
```

```python
reg = linear.fit([df['Hours_of_study']],[df['pass/fail']])
```

```python
pred = reg.predict([df['Hours_of_study']])
```

```python
plt.plot(df.Hours_of_study.values,pred.flatten())
plt.show()
```
