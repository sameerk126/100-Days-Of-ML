```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df_train = pd.read_csv('Datasets/train.csv')
```

```python
df_test = pd.read_csv('Datasets/test.csv')
```

```python
df = df_train.append(df_test)
```

```python
df.columns
```

```python
df.drop(columns = ['id', 'CustomerId', 'Surname'],inplace  = True)
```

```python
df
```

```python
from sklearn.preprocessing import LabelEncoder
label = LabelEncoder()
```

```python
# df['species']= label_encoder.fit_transform(df['species']) 
def label_enc(df,cols):
    for column in cols:
        df[column] = label.fit_transform(df[column])
```

```python
label_enc(df,['Geography', 'Gender'])
```

```python
df
```

```python
df.info()
```

```python
df['Age'] = df['Age'].astype(int)
```

```python
df.Exited.unique()
```

```python
df['Exited'] = df['Exited'].map({0. :False,  1.:True, 'nan':np.nan})
```

```python
df.describe()
```

```python
test = df[df['Exited'].isnull()]
```

```python
train = df[df['Exited'].isnull() == False]
```

```python
y = train['Exited'].astype(bool)
```

```python
y.info()
```

```python
X = train.drop('Exited',axis = 1)
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.4,random_state=1)
```

```python
from sklearn.linear_model import LogisticRegression
log_reg = LogisticRegression()
log_reg.fit(X_train,y_train)
log_reg_pred = log_reg.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
pred = log_reg_pred
print('Accurecy is ',accuracy_score(y_test,log_reg_pred))
print(confusion_matrix(y_test,log_reg_pred))
print(classification_report(y_test,log_reg_pred))
```

```python
from sklearn.svm import SVC 
svc = SVC()
svc.fit(X_train,y_train)
svc_pred = svc.predict(X_test)
```

```python
pred = svc_pred
print('Accurecy is ',accuracy_score(y_test,log_reg_pred))
print(confusion_matrix(y_test,log_reg_pred))
print(classification_report(y_test,log_reg_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf.fit(X_train,y_train)
rf_pred = rf.predict(X_test)
```

```python
pred = rf_pred
print('Accurecy is ',accuracy_score(y_test,log_reg_pred))
print(confusion_matrix(y_test,log_reg_pred))
print(classification_report(y_test,log_reg_pred))
```
