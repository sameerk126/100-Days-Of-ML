```python
df_test = df[df['Exited'].isnull()]
df_train  = df[df['Exited'].isnull() == False]
```

```python
df_train
```

```python
df_train.shape
```

```python
X = df_train.drop('Exited',axis =1)
y = df_train['Exited']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.4, random_state=1)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train_scaled = sc.fit_transform(X_train)
X_test_scaled = sc.transform(X_test)
df_test_scaled = sc.transform(df_test)
```

```python
plt.subplots(figsize = (24,9))
plt.subplot(1,2,1)
sns.boxplot(X_train)
plt.subplot(1,2,2)

sns.boxplot(X_train_scaled)
```

```python
from sklearn.linear_model import LogisticRegression
log_reg = LogisticRegression()
log_reg.fit(X_train,y_train)
log_reg_pred = log_reg.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
pred = log_reg_pred
print('Accurecy is ',accuracy_score(y_test,pred))
print(classification_report(y_test,pred))
print(confusion_matrix(y_test,pred))
```

```python
from sklearn.svm import SVC
svc = SVC()
svc.fit(X_train,y_train)
svc_pred = svc.predict(X_test)
```

```python
def print_report( test,pred):
#     pred = svc_pred
    print('Accurecy is ',accuracy_score(test,pred))
    print(classification_report(test,pred))
    print(confusion_matrix(test,pred))
```

```python
pred = svc_pred
print('Accurecy is ',accuracy_score(y_test,pred))
print(classification_report(y_test,pred))
print(confusion_matrix(y_test,pred))
```

```python
from sklearn.naive_bayes import BernoulliNB
nb = BernoulliNB()
nb.fit(X_train,y_train)
nb_pred = nb.predict(X_test)
```

```python
print_report(y_test,nb_pred)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf.fit(X_train,y_train)
rd_pred = rf.predict(X_test)
```

```python
print_report(y_test,rd_pred)
```

```python
from sklearn.ensemble import AdaBoostClassifier
ada = AdaBoostClassifier()
ada.fit(X_train,y_train)
ada_pred = ada.predict(X_test)
```

```python
print_report(y_test, ada_pred)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
gb = GradientBoostingClassifier()
gb.fit(X_train,y_train)
gb_pred = gb.predict(X_test)
```

```python
print_report(y_test,gb_pred)
```

```python
from sklearn.neighbors import KNeighborsClassifier
neigh = KNeighborsClassifier(n_neighbors=1000)
neigh.fit(X_train, y_train)
neigh_pred = neigh.predict(X_test)
```

```python
print_report(y_test, neigh_pred)
```
