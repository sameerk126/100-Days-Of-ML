```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
df = pd.read_csv('Datasets/ufc-fighters-statistics.csv')
df.shape
```

```python
df.isnull().sum()
```

```python
sns.heatmap(df.isnull())
```

```python
df['weight_in_kg'] = df['weight_in_kg'].fillna(df['weight_in_kg'].mean())
```

```python
df.drop('nickname',axis = 1,inplace = True)
```

```python
sns.boxplot(df['height_cm'])
```

```python
df['height_cm']  = df['height_cm'].fillna(df['height_cm'].quantile(.5))
```

```python
df.info()
```

```python
df['date_of_birth'] = pd.to_datetime(df['date_of_birth']).dt.year
df['date_of_birth'] = df['date_of_birth'].fillna(df['date_of_birth'].mean())
```

```python
for i in df.columns:
    print("NaN values in ",i , '  =  ',df[i].isna().sum())
```

```python
df['stance'] = df['stance'].fillna('Free')
```

```python
df.corr(method='spearman',numeric_only=True).style.background_gradient()
```
