```python
x = np.random.rand(50)
y = np.random.rand(50)
plt.scatter(x,y)
```

```python
colors = np.random.rand(50)
sizes = 1000*np.random.rand(50)
plt.scatter(x,y,c=colors,s = sizes,alpha = .9)
plt.xlabel("X - Axis")
plt.ylabel("Y - Axis")
plt.title("Random Data")
```
